```python
import pandas as pd
import os
```

```python
path = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-i-mihajlova/shared/homeworks/python_ds_miniprojects/5_subsid/subsid/'
# get all files with data
files = os.listdir(path)
```

```python
# the good sequence of the columns in dataframe
good_columns = ['SUBS_ID', 'FILIAL_ID', 'PROD_ID', 'ACT_DTTM']
```

```python
# change the sequence of columns in df and renemes it to lower, deletes rows with empty subs_id
def make_df(df):
    df = df.reindex(columns = good_columns)
    df = df.dropna(subset = ['SUBS_ID'])
    return df.rename(columns = lambda x: x.lower())
    
```

```python
# create a big data frame with all info about sales

all_sales = pd.DataFrame([])
for i in range(1,4):
    df = pd.read_csv(path + files[i], sep = ';')
    df = make_df(df)
    all_sales = pd.concat([all_sales, df], ignore_index=True)
```

```python
# adds a prefix id if necessary
def change_id(x):
    if x.startswith('id'):
        return x
    else:
        return 'id' + x
```

```python
all_sales['subs_id'] = all_sales.subs_id.apply(change_id)
```

```python
# load activation info
sales = pd.read_csv(path + files[0], sep = ';')
```

```python
# rename the columns to lowercase
sales = sales.rename(columns = lambda x: x.lower())
```

```python
# save the dates in format day-month-year hours-minuites
sales.start_dttm = pd.to_datetime(sales.start_dttm, format='%d-%m-%Y %H:%M')
sales.end_dttm = pd.to_datetime(sales.end_dttm, format='%d-%m-%Y %H:%M')
```

```python
# calculate the activation time
sales['activation_time'] = sales.end_dttm - sales.start_dttm
```

```python
# save all with long activation time
sales = sales[sales.activation_time > '5m']

```

```python
# merge both tables
sales_info = all_sales.merge(sales, on='subs_id')
```
